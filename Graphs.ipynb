```python
from matplotlib import pyplot as plt
import numpy as np
 
 
# Creating dataset
labels = ['0', '1', '2']
colors = ['gold', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0)

 #26\% zeros, 22\% ones, 51\% twos
 #40\% zeros, 17\% ones, 39\% two
im2gps3k_percentage = [26, 22, 51]
yfcc4k_percentage = [40, 17, 39]

# Creating plot
fig = plt.plot(figsize =(5, 5))
#plt.pie(im2gps3k_percentage, labels = classes)
#plt.pie(im2gps3k_percentage, labels=labels, colors=colors,
#        autopct='%1.1f%%', shadow=True, startangle=140)
patches, texts = plt.pie(im2gps3k_percentage, colors=colors, labels=labels, labeldistance=0.5, shadow=False, startangle=90, 
                         wedgeprops = {"edgecolor" : "black", 'linewidth': 1.5, 'antialiased': True})
for text in texts:
    text.set_fontweight('bold')
    text.set_horizontalalignment('center')

plt.tight_layout()
# show plot
plt.show()
```
